# Решение дифф. уравнений численными методами при помощи Python

## Система дифференциальных уравнений, представляющая собой Аттрактор Лоренца

 ![Lorenz attractor](https://wikimedia.org/api/rest_v1/media/math/render/svg/f2c661afc7030ef444c811349ec573a20a3dce61).

```python
import numpy as np
import matplotlib.pyplot as plt

from lorenz import LorenzParams, generate_initial_val, lorenz
from scipy.integrate import solve_ivp

# Извлечение генератора случайных чисел из библиотеки numpy
rng = np.random.default_rng()
```

Введём переменные со значениями параметров, присутствующих в записи Аттрактора Лоренца

```python
p = LorenzParams()
method = 'RK45'
u0 = generate_initial_val(rng)

t_eval = np.arange(p.tau_0, 2 * p.tau_0 + p.delta_tau, 0.00005)
solution = solve_ivp(lorenz, [0, t_eval[-1]], u0, args=[p], method=method, t_eval=t_eval)
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.plot(*solution.y, lw=0.5)
```
